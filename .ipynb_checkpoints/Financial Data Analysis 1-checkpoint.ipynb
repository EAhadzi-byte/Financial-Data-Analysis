```python
#section 2.2: cleaning data columns (removing whitespaces, special characters,filling spaces, among others)

def clean_column(col):
    col = col.strip()
    col = col.replace(" ", "_")
    col = col.replace("(", " ")
    col = col.replace(")", " ")
    col = col.upper()
    return col
    
#calling the function by creating a new_column
new_financials_columns = []

for c in financials.columns:
    clean_columns = clean_column(c)
    new_financials_columns.append(clean_columns)

financials.columns = new_financials_columns
print(financials.columns)
```

```python
 #Function to plot the Interactive Financial Plot
# Select Top 5 Industries by Revenue
def plot_top5_industries_revenue(): 
    top_5industries = (
    new_financials.groupby('CATEGORY')['REVENUE'].sum()
    .sort_values(ascending=False)
    .head(5).index)

    filtered_financials = new_financials[new_financials['CATEGORY'].isin(top_5industries)]

# Aggregate Revenue by Year and Industry
    industry_yearly = (
        filtered_financials
        .groupby(['YEAR', 'CATEGORY'])
        ['REVENUE'].sum()
        .reset_index())


# Create Interactive Plot
    fig = go.Figure()

    for industry in top_5industries:
        industry_data = industry_yearly[industry_yearly['CATEGORY'] == industry]
    
        fig.add_trace(go.Scatter(
            x=industry_data['YEAR'],
            y=industry_data['REVENUE'],
            mode='lines+markers',
            name=industry,
            hovertemplate=('Industry: %{text}<br>""Year: %{x}<br>""Revenue: %{y:,.0f}<extra></extra>'),
            text=[industry] * len(industry_data)))


# Post-COVID Highlight
    fig.add_vrect(
        x0=2020, x1=2023,
        fillcolor='red',
        opacity=0.1,
        layer='below',
        line_width=0,
        annotation_text='Post-COVID Period',
        annotation_position='top left')

# Layout Customization
    fig.update_layout(
        title='Interactive Revenue Trends for Top Five Industries (2009–2023)',
        xaxis_title='Year',
        yaxis_title='Total Revenue',
        template='plotly_white',
        legend_title='Industry',
        hovermode='x unified')

    fig.show()
```

```python
# Function to plot the Histogram and Box Plot
def net_income_distribution():

    # Plot the figure size
    fig, axes = plt.subplots(1, 2, figsize=(10, 7))

# Histogram with KDE
    sns.histplot(new_financials['NET_INCOME'],
    kde=True,color='green', ax=axes[0])
    plt.title('Distribution of Net Income (Histogram with KDE)')
    plt.xlabel('Net Income')
    plt.ylabel("Frequency")

#Box plot by Category
    sns.boxplot(
        x="CATEGORY",
        y='NET_INCOME',hue='CATEGORY',
        data=new_financials,
        palette='Set2',legend= False)

#Adding Titles and labels
    plt.title('Net Income by Industry Category')
    plt.xlabel('Industry Category')
    plt.ylabel('Net Income')
    plt.xticks(rotation= 90)

#show the plot
    plt.tight_layout()
    return fig
```

```python
# Function to plot the Joint Scatter plot
def scatter_plot():
    g = sns.jointplot(
        data=new_financials,
        x='DEBT/EQUITY_RATIO',
        y='NET_PROFIT_MARGIN',
        hue='CATEGORY')
    plt.subplots_adjust(top=0.9)
    g.fig.suptitle('Risk (Debt/Equity) vs Profit Margin')
    return g.fig
```

```python
# Function to plot the Violin
#Create COVID period column using Violin plot
def violin_plot_covid():
    new_financials.loc[:,'Period'] = new_financials['YEAR'].apply(lambda x: 'Pre-COVID' if x < 2020 else 'Post-COVID')

#Create Violin plot
    fig, ax = plt.subplots(figsize=(8, 5))
    sns.violinplot(data=new_financials,x='Period'
               ,y='NET_INCOME', hue='Period'
               ,palette='muted' , legend= False)

    plt.title('Net Income Distribution: Pre vs Post COVID')
#show violinplot
    return fig   
```

```python
# Function to plot the Correlation Heatmap
def  correlation_heatmap():
    corr = new_financials[['REVENUE', 'NET_INCOME','MARKET_CAP_IN_B_USD', 'ROE',
                       'ROA', 'DEBT/EQUITY_RATIO', 'INFLATION_RATE_IN_US']].corr()
    
#Create the heatmap
    fig, ax = plt.subplots(figsize=(8, 6))
    sns.heatmap(corr, annot=True, fmt='.2f', cmap='YlGnBu',cbar=True)

# Titles
    plt.title('Correlation Heatmap of Financial Metrics', fontsize=12)

#Display the plot
    return fig
```

```python
#Function to plot the Interactive bubble chat
#indication post_covid Era
def Interactive_bubble_plot():
    post_covid = new_financials[new_financials["YEAR"] >= 2020]

#Create an interactive scatter plot
    fig = px.scatter(
        post_covid,
        x='INFLATION_RATE_IN_US',
        y="ROI",
        color="CATEGORY",
        size='MARKET_CAP_IN_B_USD',
        hover_name="COMPANY",
        title="Inflation Impact on ROI (Post-COVID)",
        color_discrete_sequence=px.colors.qualitative.Safe)

    fig.update_layout(
        title_font_size=18,
        hovermode="closest",
        showlegend=True,
        legend_title_text="CATEGORY")

    fig.show()
```

Section 3.5 Data Aggregation and Grouping
In this Section I will be doing summary statistics for the data using groupby(),Sum(),mean() which will help us to discover patterns in revenue,profitability,employment and financial performance across industries,companies and years.

```python
# Average Revenue Per Company
Avg_revenue_company = new_financials.groupby("COMPANY")["REVENUE"].mean().round(2).sort_values(ascending=False)
Avg_revenue_company.head()
```

```python
#Total Market Cap by Category(Industry)
Market_by_Category =financials.groupby("CATEGORY")['MARKET_CAP_IN_B_USD'].sum().sort_values(ascending=False)
Market_by_Category.head()
```

```python
#Average employment by Company
Ave_employees = new_financials.groupby("COMPANY")["NUMBER_OF_EMPLOYEES"].mean().round(0).sort_values(ascending=False)
Ave_employees.head()
```

```python
#Industry Growth in Net Income after post covid_era(2020 - 2023)
Net_income_trend = post_covid.groupby("CATEGORY")["NET_INCOME"].sum().round(2).sort_values(ascending=False)
Net_income_trend.head()
```

```python
# Companies with Highest Debt to Equity Ratio(Risk Indicator)
high_risk = new_financials[["COMPANY", "DEBT/EQUITY_RATIO"]].sort_values(by = "DEBT/EQUITY_RATIO",ascending=False)
high_risk.head(5)
```
